# `**BÀI 1: Đổi mới**`

1. **Abstraction**: Số cây đốn hạ của 2 đội lần lượt là a và b, ngày nghỉ của 2 đội lần lượt là thứ k và thứ m. Tìm số ngày sau khi đốn hết n cây.
2. **Recognition**: sử dụng thuật toán binary search
3. **Algorithm designed**: (Lưu đồ thuật toán)
4. **Alogrithm Complexity**: O(n)

5. **Programming**:

```python
a,k,b,m,n = [int(i) for i in input().split()]
low = n//(a+b)
high = n
while low <= high:
    mid = (low+high)//2
    if a*(mid-mid//k)+ b*(mid-mid//m) >= n:
        day = mid
        high = mid - 1
    else:
        low = mid + 1
print(day)
```
